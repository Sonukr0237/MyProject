```python
import pandas as pd
import re
df = pd.read_csv('mydata.csv')
df
```

```python
# Drop duplicates

df = df.drop_duplicates(subset = 'Index')
df
```

```python
df.info()
```

```python
# Validation in phone number
def cleanPhoneNumber(phone):
    cleaned = re.sub(r'\D', '', phone)
    return cleaned
df['cleanedPhone'] = df['Phone 1'].apply(cleanPhoneNumber)
# Check if the cleaned phone numbers contain only digits
df['only digit'] = df['cleanedPhone'].str.isdigit()
df_result = df[['Phone 1', 'cleanedPhone', 'only digit']]
df_result
```

```python
# It should have 10 digits
df['phoneLength'] = df['cleanedPhone'].str.len()
df['ten digit'] = df['phoneLength'] == 10
dfResult = df[['Phone 1', 'cleanedPhone', 'only digit', 'phoneLength', 'ten digit']]
dfResult
```

```python
# Implementation ofcondition for showing ten digits phone no data
dfResult = df.loc[(df['phoneLength'] == 10) & (df['only digit'] == True)]
dfResult
```

```python
dfResult = df.drop(columns=['only digit', 'ten digit'])
dfResult
```
