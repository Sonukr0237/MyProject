```python
import pandas as pd
walmart = pd.read_csv('Walmart.csv')
walmart
```

```python
# find max and min of the data
max = walmart['Weekly_Sales'].max()
min = walmart['Weekly_Sales'].min()
print(max)
print(min)
```

```python
# Lower Extreme and Upper Extreme:
# Outliers are data points that are more extreme than than Q1 - 1.5 * IQR or Q3 + 1.5 * IQR.

# Lower Extreme : Q1 - 1.5 * IQR

# Upper Extreme : Q3 + 1.5 * IQR

# lets's find Q1 and Q3

import numpy as np
q1, q3 = np.percentile(walmart['Weekly_Sales'], [25, 75])
print(q1, q3)
```

```python
# let's find interquartile range
iqr = q3 -q1
iqr
```

```python
# let's find lower and upper extreme
lx = q1 - 1.5 * iqr
ux = q3 + 1.5 * iqr
print(lx, ux)
```

```python
outliersPresent = lx > min
outliersPresent
```

```python
outliersPresent = ux < max
outliersPresent
```

```python
# Remove Outliers 
noOutliers = walmart.loc[(walmart['Weekly_Sales'] < ux) & (walmart['Weekly_Sales'] > lx)]
noOutliers
```
