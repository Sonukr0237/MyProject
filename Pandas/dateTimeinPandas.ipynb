```python
import pandas as pd

df = pd.read_csv('Walmart.csv')
df
```

```python
df.info()
```

```python
# Convert columns in date time format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y', errors='coerce')
df
```

```python
df.info()
```

```python
# Filtering in date time
df.loc[df['Date']>'2010-01-01']
```

```python
df.loc[df["Date"]>'2010']
```

```python
df.loc[df['Date']>='2010-05']
```

```python
df.loc[df["Date"]=='2012-10-12']
```

```python
# Multiple condition in date
edf = df.loc[(df['Date']>='2010') & (df['Date']<'2011-01-01')]
edf
```

```python
# sort Date
edf.sort_values('Date')
```

```python
# set date time as Index
edf = df.set_index('Date')
edf
```

```python
# If we have set date as index, can perform filtering very easily 
edf['2010':'2011']
```

```python
# Get day from date
df['day'] = df['Date'].dt.day
df[['Date', 'day']]
```

```python
# get month from a month
df['month'] = df['Date'].dt.month
df[['Date', 'month']]
```
