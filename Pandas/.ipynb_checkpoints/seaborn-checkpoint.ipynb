```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# Creating first graph
sns.load_dataset('tips')
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', data = df)
plt.show()
```

```python
# hue : This parameter take column name for colour encoding
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = df)
plt.show()
```

```python
# Pallete : if want to change the color encoding smokers and non-smokers
c = {'No':'Green', 'Yes':'Red'}
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'smoker', hue_order = ['No', 'Yes'], palette = c, data = df)
plt.show()
```

```python
sns.scatterplot(x = 'total_bill', y = 'tip', hue = 'size', data = df)
plt.show()
```

```python
# relplot: relplot is used for many purposes, but the main purpose behind using relplot is sub-plotting. Suppose we want to subplot our graph base on smoker and non-smoker
# we can use row, col to achieve this result
sns.relplot(x='total_bill', y='tip', data=df, col='smoker')
plt.show()
```

```python
# If using row
sns.relplot(x='total_bill', y='tip', data=df, row='smoker')
plt.show()
```

```python
# Using size
%matplotlib notebook
sns.relplot(x='total_bill', y='tip', hue='size', data=df, col='size')
plt.show()
```

```python
%matplotlib inline
sns.relplot(x='total_bill', y='tip', hue='size', data=df, col='size', col_wrap = 3)
plt.show()
```

```python
# Using catagorical column
sns.relplot(x = 'total_bill', y = 'time', data = df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('flights')
df
```

```python
# want to know how many passengers on year
sns.relplot(x= 'passengers', y = 'month', hue = 'month', data = df)
plt.show()
```

```python
# catplot
df = sns.load_dataset('tips')
sns.catplot(x = 'time', y = 'tip', data= df)
plt.show()
```

```python
sns.catplot(x = 'total_bill', y = 'day', data= df)
plt.show()
```

```python
# change the graph by bar
sns.catplot(x = 'total_bill', y = 'day', data= df, kind= 'bar')
plt.show()
```

```python
sns.catplot(x = 'total_bill', y = 'day', data= df, kind = 'point')
plt.show()
```

```python
sns.catplot(x = 'total_bill', y = 'day', data= df, kind = 'violin')
plt.show()
```

```python
sns.catplot(x = 'day', y = 'total_bill', data= df, kind='boxen')
plt.grid()
plt.show()
```

```python
# countplot : method is used to show the counts of observations in each categorical bin
sns.countplot(y='smoker', data=df)
```

```python
df = pd.read_csv('Pokemon.csv')
sns.countplot(y = 'Type 1', data = df)
plt.show()
```

```python
# pairplot
df = sns.load_dataset('tips')
sns.pairplot(data = df, hue = 'sex')
plt.show()
```
