```python
# Standard Deviation
dogs = ['d1', 'd2', 'd3', 'd4', 'd5']
heights = [600, 470, 170, 430, 300]

mean = sum(heights)/len(dogs)
mean
```

```python
# find variance 
differences = [height - mean for height in heights]
squaredDifferences = [diff ** 2 for diff in differences]
variance = sum(squaredDifferences)/len(squaredDifferences)
variance
```

```python
# Standard deviation is just square root of variance
standardDeviation = math.sqrt(variance)
standardDeviation
```
