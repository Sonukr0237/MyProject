```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objs as go
items = [10,5,7,9,20,6,11]
days = ['Day1', 'Day2', 'Day3', 'Day4', 'Day5', 'Day6', 'Day7']

plt.plot(days, items)
plt.show()
```

```python
# Properties 
plt.plot(days, items, color='green', linestyle='--')
plt.title('Shop-Data')
plt.xlabel('Days')
plt.ylabel('Items')
plt.grid()
plt.show()
```

```python
# Markers
plt.plot(days, items, color='green', marker='o')
plt.title('Shop-Data')
plt.xlabel('Days')
plt.ylabel('Items')
plt.grid()
plt.show()
```

```python
# The problem of overlapping and how to overcome it
data1 = [10,9,5,7,6,11,20]
data2 = [5,9,2,5,6,8,2]

plt.plot(data1,data2,marker = 'o')
plt.title('Data')
plt.xlabel("Data 1")
plt.ylabel("Data 2")
plt.grid()
plt.show()
```

```python
data1 = [10,9,5,7,6,11,20]
data2 = [5,9,2,5,6,8,2]
data2.sort()

plt.plot(data1,data2,marker = 'o')

plt.title('Data')
plt.xlabel("Data 1")
plt.ylabel("Data 2")
plt.grid()
plt.show()
```
